```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the text file
data = pd.read_csv('sequential-time.txt', sep="\t", header=None)

# Plot the data
plt.scatter(data[1], data[0])
plt.xlabel('Amount of values processed')
plt.ylabel('Time(s)')
plt.title('Time taken to process a # of values sequential')
plt.axhline(y=0.01, color='r', linestyle='-')
plt.axvline(x=350000, color='r', linestyle='-')
plt.show()
```

```python
# Read the data from the text file
data = pd.read_csv('parallel-time.txt', sep="\t", header=None)

# Plot the data
plt.scatter(data[1], data[0])
plt.xlabel('Amount of values processed')
plt.ylabel('Time(s)')
plt.title('Time taken to process a # of values parallel')
plt.ylim = (0, 0.05)
plt.axhline(y=0.001, color='r', linestyle='-')
plt.axvline(x=350000, color='r', linestyle='-')
plt.show()
```
